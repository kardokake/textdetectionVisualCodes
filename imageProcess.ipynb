```python
# import modules

import time
import os
import matplotlib.pyplot as plt
from scipy.misc import imread, imsave, imshow
from skimage.measure import label as skimage_label
from skimage.draw import line as skimage_line
from skimage.draw import set_color as skimage_set_color
```

```python
# define functions 

def fileNameLink(filename, dirlist):
    # use detection hotmap file name to find corresponding bounding box filename
    n1 = filename.split('_')[-1]
    for filename in dirlist:
        if filename.endswith('.png'):
            fstring = filename.split('.')[0]
            n2 = fstring.split('_')[1]
            n3 = fstring.split('_')[-1]
            if n2 == n1:
                return 'rct_img_'+n3+'.png'
    return 

def fileNameLink1(filename, dirlist):
    # use detection hotmap file name to find corresponding original rgb image filename
    n1 = filename.split('_')[-1]
    for filename in dirlist:
        if filename.endswith('.png'):
            fstring = filename.split('.')[0]
            n2 = fstring.split('_')[1]
            n3 = fstring.split('_')[-1]
            if n2 == n1:
                return 'img_'+n3+'.jpg'
    return 

def imgShow(image, title):
    plt.imshow(image)
    plt.axis('off') # clear x- and y-axes
    plt.title(title)
    plt.show()
    return

def imgAlpha(image1, image2):
    plt.imshow(image1)
    plt.imshow(image2, alpha = 0.5)
    plt.axis('off') # clear x- and y-axes
    plt.show()
    return

def saveMovie(frames):
    import matplotlib
    import matplotlib.pyplot as plt
    import matplotlib.animation as manimation

    FFMpegWriter = manimation.writers['ffmpeg']
    metadata = dict(title='Image Thresholding', artist='Siyu Zhu',
        comment='Cool!')
    writer = FFMpegWriter(fps=10, metadata=metadata)

    fig = plt.figure()
    with writer.saving(fig, "imageThresholding.mp4", 100):
        for frame in frames:
            ax = fig.add_subplot(111)
            ax.set_aspect(1)
            plt.axis('off')
            res = ax.imshow(frame, cmap=plt.cm.hot, interpolation='nearest')
            writer.grab_frame()
            
def imgNorm(img):
    img = numpy.float32(img)
    img = img - img.min()
    img = img / img.max()
    img = img * 255
    img = numpy.uint8(img)
    return img

def grey2rgb(canvas):
    if len(canvas.shape) == 3 and canvas.shape[2] == 3:# if color image
        pass
    elif len(canvas.shape) == 2: # if grayscale image
        colorcanvas = numpy.empty((canvas.shape[0], canvas.shape[1], 3))
        colorcanvas[:, :, 0] = canvas
        colorcanvas[:, :, 1] = canvas
        colorcanvas[:, :, 2] = canvas
    return colorcanvas

def setcolor(img, coor, color, line_width = 1):
    oimg = img
    xcor, ycor = coor
    coor = map(lambda x, y: (x, y), xcor, ycor)
    if len(img.shape) == 3 and img.shape[2] == 3 and len(color) == 3:
        #print 'color image'
        for x, y in coor:
            for i, k in enumerate(color):
                if line_width == 1:
                    img[x, y, i] = k
                else:
                    for w in range(line_width):
                        img[x+w, y+w, i] = k
                        img[x-w, y-w, i] = k
                
        return img
                
    elif len(img.shape) == 2 and len(color) == 1:
        #print 'greyscale image'
        k = color[0]
        for x, y in coor:
            if line_width == 1:
                img[x, y, i] = k
            else:
                for w in range(line_width):
                    img[x+w, y+w, i] = k
                    img[x-w, y-w, i] = k
        return img
            
    else:
        print 'Color length must match the image type!'
        return

def drawbb(canvas, Coor, mycolor, linewidth):
       
    for coor in Coor:
        x1 = coor[0]
        y1 = coor[1]
        x2 = coor[2]
        y2 = coor[3]
        line1 = skimage_line(x1, y1, x2, y1)
        line2 = skimage_line(x1, y1, x1, y2)
        line3 = skimage_line(x1, y2, x2, y2)
        line4 = skimage_line(x2, y1, x2, y2)
        canvas = setcolor(canvas, line1, mycolor, linewidth)
        canvas = setcolor(canvas, line2, mycolor, linewidth)
        canvas = setcolor(canvas, line3, mycolor, linewidth)
        canvas = setcolor(canvas, line4, mycolor, linewidth)
    return canvas

def image2bb(image, thre):
    lbimg = skimage_label(numpy.uint8(image > thre), background = 0)
    string = []
    for l in xrange(lbimg.max()+1):     
        x, y = numpy.where(lbimg == l)  
        x0 = min(x)  
        y0 = min(y)    
        x1 = max(x)       
        y1 = max(y)  
        bb = [y0, x0, y1, x1]  
        bb = map(str, bb)   
        string.append(','.join(bb))  
    string = '\r\n'.join(string)   
    return string    

def labelStr2coor(label_string):
    ''' correct bounding box format for prediction files '''
    Coor = []
    for bb in label_string.split('\r\n'):
        if not bb.strip(): # in case that label string contains nothing
            continue
        coor = map(int, bb.split(','))
        c = list(coor)
        c[0] = coor[1]
        c[1] = coor[0]
        c[2] = coor[3]
        c[3] = coor[2]
        Coor.append(c)
    return Coor

# threshold for movie making
T0 = numpy.arange(0.0, 0.5, 0.01)
T1 = numpy.arange(0.5, 0.8, 0.005)
T2 = numpy.arange(0.8, 1.0, 0.01)
THRE = numpy.concatenate((T0, T1, T2))

def frameMaker(img, THRE):
    frame = []
    for thre in THRE:
        im = img < thre
        strlabel = image2bb(im, thre)
        Coor = labelStr2coor(strlabel)
        im = drawbb(im, Coor, [0], 5)
        frame.append(im)
    return frame

def loadScaleImage(subdirs_full):
        # find all scales detection hotmap and put into scaleContainer
        subdirlist = os.listdir(subdirs_full)
        for i, filename in enumerate(subdirlist):
            image = plt.imread(os.path.join(subdirs_full, filename))
            if i == 0:
                scaleContainer = image
            else:
                scaleContainer = numpy.dstack((scaleContainer, image))
        return scaleContainer
```
