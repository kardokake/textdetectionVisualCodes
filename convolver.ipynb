```python
psize = 32
ssize = 16
para1 = 0.3
para0 = 0.8
```

```python
def image2label(img, bb, psize = 32, ssize = 16, para1 = 0.3, para0 = 0.8):
    ind1, ind2 = map(lambda x:range(0, x-psize, ssize), img.shape)
    for b in bb:
        labelMat = numpy.zeros((len(ind1), len(ind2)), dtype = numpy.uint8)
        if not size_comp(b, psize, para1):
            continue
        for p, i in enumerate(ind1):
            for q, j in enumerate(ind2):
                if overlap_comp(b, [i, j, psize, psize], para0):
                    print b
                    labelMat[p, q] = 1
    return labelMat
```

```python
def size_comp(bb, psize, para):
    y1, x1, y2, x2 = bb
    w = x2 - x1
    h = y2 - y1
    if (psize < w*(para+1) and psize > w*(1-para)) or (psize < h*(para+1) and psize > h*(1-para)):
        return True
    return False

def overlap_comp(bb1, bb2, para):
    y1, x1, y2, x2 = bb1
    x3, y3, w3, h3 = bb2
    a = (min(x2, x3+w3) - max(x1, x3))
    b = (min(y2, y3+h3) - max(y1, y3))
    if (a > 0 and b > 0 and a*b > w3 * h3 * para):
        return True
    return False
```

```python
from scipy.misc import imresize
resize_factor = 0.95 ** numpy.arange(0, 30)
for i in resize_factor:
    newimg = imresize(img, i)
    newbb = numpy.uint16(bb*i)
    labelimg = image2label(newimg, newbb, psize = 16, ssize = 4, para0 = 0.8, para1 = 0.3)
    labelimg = imresize(labelimg, img.shape, interp = 'nearest')
    plt.imshow(labelimg)
    plt.show()
```
