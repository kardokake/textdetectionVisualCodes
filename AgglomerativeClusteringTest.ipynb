```python
from sklearn.cluster import AgglomerativeClustering, MiniBatchKMeans  
import matplotlib.pyplot as plt
import time
```

```python
n = 1000
m = 100
linkage = 'average'
affinity = 'cosine'
data = numpy.random.rand(n, m)
et1 = []
et2 = []
N_clusters = range(100, 1100, 100)
for n_clusters in N_clusters:
    print 'No. of clusters', n_clusters
    model1 = AgglomerativeClustering(linkage = linkage, affinity = affinity, n_clusters = n_clusters, connectivity = None)
    model2 = MiniBatchKMeans(n_clusters = n_clusters)

    time0 = time.time()
    model1.fit(data)
    et1.append(time.time()- time0)

    time0 = time.time()
    model2.fit(data)
    et2.append(time.time() - time0)
```

```python
# compute the precision, recall and f measure for testing
fig, ax = plt.subplots(figsize=(5, 5))
plt.plot(et1, label = 'minibatchKmeans')
plt.plot(et2, label = 'agglomerative')
plt.grid(True)
plt.xlabel('No. of Clusters')
plt.ylabel('Elapse time (s)')
plt.title(r'Relationship of speed and No. of clusters')
ax.legend(loc = 1)
```

```python
from sklearn.cluster import AgglomerativeClustering, MiniBatchKMeans
import scipy.cluster.hierarchy as hac
import matplotlib.pyplot as plt
import time
```

```python
k_ = 10 # no. of groups
m_ = 400 # no. of samples in each group
f_ = 2 # no. of features for each sample

X = numpy.zeros((0, f_))
for i in xrange(k_):
    x = np.random.multivariate_normal(numpy.random.rand(f_)*100, numpy.random.rand(f_, f_)*33, m_)
    X = numpy.concatenate((X, x))
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
p1 = plt.scatter(X[:, 0], X[:, 1])
plt.grid(True)
plt.xlabel('Iterations')
plt.ylabel('Error')
plt.title(r'Testing curve')
#ax.legend(loc = 1)
#plt.savefig('error.png')
#plt.savefig('error.pdf')
```

```python
linkage = 'ward'
affinity = 'euclidean'
n_clusters = k_
data = X
```

```python
t = time.time()
model1 = AgglomerativeClustering(linkage = linkage, affinity = affinity, connectivity = None, n_clusters = n_clusters)
label1 = model1.fit(data).labels_
t1 = time.time() - t

t = time.time()
model2 = MiniBatchKMeans(n_clusters = n_clusters)
label2 = model2.fit(data).labels_
t2 = time.time() - t

t = time.time()
model3 = hac.linkage(data, method = linkage, metric = affinity)
label3 = hac.fcluster(model3, n_clusters, 'maxclust')
t3 = time.time() - t
```

```python
def label_plot(label, cmap, title):
    fig, ax = plt.subplots(figsize=(5, 5))
    for i, l in enumerate(set(label)):
        x = data[label == l]
        p1 = plt.scatter(x[:, 0], x[:, 1], color = cmap[i])
        plt.grid(True)
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.title(title)
```

```python
cmap = numpy.random.rand(n_clusters, 3)    
label_plot(label1, cmap, 'Agglomerative ElapseTime %.2f s'%t1)
label_plot(label2, cmap, 'Kmeans ElapseTime %.2f s'%t2)
label_plot(label3, cmap, 'Agg_scipy ElapseTime %.2f s'%t3)
```
