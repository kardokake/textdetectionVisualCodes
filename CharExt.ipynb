```python
import detector
import parseData
from imgOp import image_load, image2edge, image_norm
from scipy.misc import imresize
from para_featExt import maximalResize, conv2d
from fileOp import codebook_load, data_load

codebook = codebook_load('codeBook.npy')
imdir = '../icdar2013/task21_22/train/image'
bbdir = '../icdar2013/task21_22/train/char_label/'
myparser = parseData.parseChar2013()
gt = myparser.parseData(imdir, bbdir)
nob = 3
fea = []
for iterNum, g in enumerate(gt):
    fname, coor = g
    img = image_load(fname, imdir)
    m, n = img.shape
    for c in coor:
        if c:
            x, y, w, h = c
            if h > w:
                padding = int(h - w)/2
                x1 = max(0, x-padding)
                x2 = min(m, x+w+padding)
                y1 = y
                y2 = y+h
            else:
                padding = int(w - h)/2
                y1 = max(0, y-padding)
                y2 = min(n, y+h+padding)
                x1 = x
                x2 = x+w
            p = img[x1:x2, y1:y2]
            p = imresize(p, (32, 32), 'bilinear')
            p = image2edge(p)
            
            f = numpy.empty((codebook.shape[0], nob, nob))
            for i, cb in enumerate(codebook):
                f[i, ...] = maximalResize(image_norm(conv2d(p, cb)), nob)
            fea.append(numpy.transpose(f, (1, 2, 0)))
    #print iterNum, len(gt)
            
            #plt.imshow(p.transpose())
            #plt.axis('off')
            #plt.show()
            #raw_input()
feaArray = numpy.asarray(fea)
feaArray = numpy.reshape(feaArray, (feaArray.shape[0], feaArray.shape[1]*feaArray.shape[2]*feaArray.shape[3]))
print feaArray.shape
numpy.save('foreArray', feaArray)
dirname = '../data2/icdar2013word/'
p, l = data_load(dirname)
ind = l == 0
p0 = p[ind, ...]
numpy.save('backArray', p0)
```

```python
foreArray = numpy.load('foreArray.npy')
forelabel = numpy.ones(len(foreArray))
backArray = numpy.load('backArray.npy')
backlabel = numpy.zeros(len(backArray))
backArray = backArray[:len(foreArray), ...]
backlabel = backlabel[:len(foreArray), ...]
fea = numpy.concatenate((foreArray, backArray))
lab = numpy.concatenate((forelabel, backlabel))
```

```python
from classifier import classifier
clfname = 'testClassifier.pkl'
clfdir = '.'
myClassifier = classifier('adaboost')
myClassifier.feature = fea
myClassifier.label = lab
myClassifier.clf_train()
myClassifier.clf_save(clfname, clfdir)
```
