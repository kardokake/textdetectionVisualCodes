```python
import pickle
import matplotlib.pyplot as plt
```

```python
def grid_plot(fi):
    ''' grid_plot(fi)
    
    plot the feature weight accross the image patches, 
    fi: 9 by 9 by 1000, numpy array, feature importance '''
    f1 = numpy.sum(fi, axis = 2)
    fig = plt.figure()
    plt.clf()
    ax = fig.add_subplot(111)
    ax.set_aspect(1)
    res = ax.imshow(f1, cmap=plt.cm.hot, interpolation='nearest')
    cb = fig.colorbar(res)
    plt.title('feature importance across grids')
    return

def codebook_plot(fi):
    ''' codebook_plot(fi)
    
    plot the feature weight across different feature banks,
    fi: 9 by 9 by 1000, numpy array, feature importances '''
    fig = plt.figure(figsize = (5,5))
    ax = fig.add_subplot(111)
    
    f2 = numpy.sum(numpy.sum(fi, axis = 0), axis = 0) # summation along codebook dimension
    f2 = numpy.sort(f2)[::-1] # sort array decending
    f3 = numpy.cumsum(f2) # accumulated distribution
    
    plt.plot(f2*100, label = '% pdf')
    plt.plot(f3, label = 'cdf')
    
    plt.grid(True)
    plt.xlabel('codebooks')
    plt.ylabel('importances')
    ax.legend(loc = 1)
    plt.title('feature importance across codebooks (sorted)')
    return
```

```python
# load classifier
clfname = '../clf2/icdar2013wordadaboost'
clf = pickle.load(open(clfname))

# generate feature importances
fi = clf.feature_importances_
feature_shape = [9, 9, 1000]
fi = numpy.reshape(fi, feature_shape)

# plot feature importances
grid_plot(fi)
codebook_plot(fi)
```
