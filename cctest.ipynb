```python
import numpy
import matplotlib.pyplot as plt

import sys
sys.path.insert(0, '../python')
%matplotlib inline
%load_ext autoreload
%autoreload 2
import os

from fileOp import pickle_load
from scipy.ndimage.morphology import binary_fill_holes
from imgOp import image_load
from skimage.filters import threshold_otsu
```

```python
dirname = '../data/ex3/finMap/'
dirlist = os.listdir(dirname)
for filename in dirlist:
    if not filename.startswith('cc'):
        continue
    if not filename.endswith('180'):
        continue
    print filename
    fd_name = 'fd'+filename[2:]+'.png'
    cc = pickle_load(filename, dirname)
    fd = image_load(fd_name, dirname)
    plt.imshow((fd.transpose()))
    plt.axis('off')
    plt.show()
    
    if not cc:
        continue
    im0 = []
    im1 = []
    im2 = []
    
    if cc[0]:
        im0 = numpy.zeros(cc[0][0].shape)
        for i, c in enumerate(cc[0]):
            #plt.imshow(c.transpose())
            #plt.show()
            #raw_input()
            if c.sum() > 0:
                im0 = im0+c*i
        #im0 = im0.transpose() > 0
        #im0 = binary_fill_holes(im0)
        plt.imshow(im0.transpose())
        plt.title('keep')
        plt.axis('off')
        plt.show()
    if cc[1]:
        im1 = numpy.zeros(cc[1][0].shape)
        for c in cc[1]:
            #plt.imshow(c.transpose())
            #plt.show()
            #raw_input()
            im1 = im1+c
        plt.imshow(im1.transpose() > 0)
        plt.title('precision discard')
        plt.axis('off')
        plt.show()   
    if cc[2]:
        im2 = numpy.zeros(cc[2][0].shape)
        for c in cc[2]:
            #plt.imshow(c.transpose())
            #plt.show()
            #raw_input()
            im2 = im2+c   
        plt.imshow(im2.transpose() > 0)
        plt.title('recall discard')
        plt.show()
    #raw_input()
    
```
