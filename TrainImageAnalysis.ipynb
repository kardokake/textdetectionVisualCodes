```python
import os
from scipy.misc import imread
import matplotlib.pyplot as plt
```

```python
imgDir = '../icdar2013/task21_22/test/image/'
dirlist = os.listdir(imgDir)
```

```python
M = []
N = []
for f in dirlist:
    filename = os.path.join(imgDir, f)
    img = imread(filename, flatten = True)
    M.append(img.shape[0])
    N.append(img.shape[1])
```

```python
# show the histogram of No. of samples contained by each cluster
fig = plt.figure(figsize = (5, 5))
n, bins, patches = plt.hist(M, 100)
n, bins, patches = plt.hist(N, 100)
plt.grid(True)
plt.xlabel('Size')
plt.ylabel('No. of Images')
plt.title('Width and Height Histogram')
```

```python
fig = plt.figure(figsize = (5, 5))
plt.plot(sorted(M))
plt.plot(sorted(N))
plt.grid(True)
plt.xlabel('image No. ')
plt.ylabel('Size')
plt.title('Width and Height plot')
```

```python
fig = plt.figure(figsize = (5, 5))
plt.boxplot((M, N), widths = 0.5)
plt.grid(True)
plt.title('Box plot of width and height of images')
```
