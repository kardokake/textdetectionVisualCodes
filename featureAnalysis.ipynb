```python
import os
import time
from sklearn.cross_validation import ShuffleSplit as shuffleSplit
```

```python
def PL_load(dirname):
        ''' load feature, label pairs from data direcotry

        dirname: string, data directory name
        Return: tuple of numpy arrays, feature, label pairs. '''
        npylist = os.listdir(dirname)
        estilen = len(npylist)
        L = []
        P = []
        for k, featname in enumerate(npylist):
                if k > 50:
                    break
                if featname.startswith('P'):
                        labename = 'L'+featname[1:]
                        feat = numpy.load(os.path.join(dirname, featname))
                        labe = numpy.load(os.path.join(dirname, labename))
                        if L == []:
                                P = feat 
                                L = labe
                        else:
                                P = numpy.concatenate((P, feat))
                                L = numpy.concatenate((L, labe))
        return P, L 
    
def featVar(P, tsize = 0.25, n_iter = 10):
    ''' featVar(P, tsize = 0.25, n_iter = 10)
    
    find the variance across different features,
    The variance is computed on a subset of samples, 
    The subset is draw from the original sample set randomly with replacement
    
    P: number array, n of samples by n of features,
    tsize: ratio of subset draw from sample matrix P, 1 for all samples from P,
    n_iter: Number of times to draw samples, the final variance is computed as average of these trials 
    Return: numpy array, variance, 1 by no_of_features. '''
    
    #tsize = 0.25 # test sample ratio
    #n_iter = 10
    n_sample, n_feat = P.shape # m, no. of samples, n, number of features
    rs = shuffleSplit(len(P), n_iter, test_size = tsize) # randomly split train and test samples
    variance = numpy.empty((n_iter, n_feat))
    i = 0
    for train, test in rs:
        P1 = P[test, ...]
        variance[i, :] = numpy.var(P1, axis = 0)
        i += 1
    variance = numpy.average(variance, axis = 0)
    return variance
```
